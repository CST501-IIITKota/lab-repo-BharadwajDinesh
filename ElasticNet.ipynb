```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.metrics as metrics
from sklearn.linear_model import ElasticNetCV
from sklearn.model_selection import GridSearchCV
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
X = train_data.iloc[:, 1:-1]
y = train_data.iloc[:, -1]
```

## ElasticNetCV

```python
alphas = np.logspace(-4, 3, 100)   # from 1e-4 to 1e3 (100 values)

# Wider search for l1_ratio (linear scale between 0 and 1)
l1_ratios = np.linspace(0.0, 1.0, 11)
```

```python
model = ElasticNetCV(
    l1_ratio=l1_ratios,
    alphas=alphas,
    cv=5,
    random_state=42,
    max_iter=10000  # increase iterations for convergence
)
```

```python
model.fit(X, y)
```

```python
X_test = test_data.iloc[:, 1:]
predictions = model.predict(X_test)
```

```python
predictions
```

```python
submission_df = pd.DataFrame({'id': test_data['id'], 'BeatsPerMinute': predictions})
display(submission_df.head())
```
